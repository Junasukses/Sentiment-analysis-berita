```python
import re
import string

def cleansing(text):
    text = re.sub(r"\d+", "", text) #angka
    text = re.sub (r'\n', '', text) #tab
    text = text.encode('ascii', 'replace').decode('ascii') #ascii decode
    text = text.translate(str.maketrans(' ', ' ', string.punctuation)) #punctuation
    text = re.sub(r"\b[a-zA-Z]\b", "", text) #satu abjad 
#     text = re.sub(r'(.)\1+',r'\1',  text)
#     text = re.sub('@[^\s]+','',text)
#     text = re.sub(r'[^\w]', ' ', text)
    return text

data['Judul Berita'] = data['Judul Berita'].apply(cleansing)
data['Berita'] = data['Berita'].apply(cleansing)
data['Judul Dan Isi'] = data['Judul Dan Isi'].apply(cleansing)
```

```python
data['Judul Berita'] = data['Judul Berita'].str.lower()
data['Berita'] = data['Berita'].str.lower()
data['Judul Dan Isi'] = data['Judul Dan Isi'].str.lower()
data
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import swifter

def preprocess(text):
    #token
    text = word_tokenize(text)
#     print(text)
    
    #stopword
    list_stopwords = stopwords.words('indonesian')
    text = [word for word in text if word not in list_stopwords]
#     print("Stopword:",text)
    
    #stem
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    text = [stemmer.stem(word) for word in text]
#     print("Stemming:",text,"\n")
    
    #mengembalikan ke text
    text = " ".join(t for t in text)
    return text

%time data['preprocess'] = data['Judul Berita'].apply(preprocess)
```

```python
from sklearn.model_selection import KFold, StratifiedKFold
from sklearn import svm
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import numpy as np

# X = np.array([[1, 2], [3, 4], [1, 2], [3, 4]])
# y = np.array([1, 2, 3, 4])
kf = StratifiedKFold(n_splits=10)

x = matrix
y = data["Label"]
q = 1
c = [1, 10, 100]
gm = [0.1, 0.005]

for train_index, test_index in kf.split(x,y):
    x_train, x_test = x[train_index], x[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
#     print(y_test)
#     print("Fold-", i, "\nTRAIN:", train_index, "\nTEST:", test_index, "\n")
    for i in range(len(c)):
        for j in range(len(gm)):
            svc = svm.SVC(kernel='sigmoid', C=c[i], gamma=gm[j]).fit(x_train, y_train)
            y_pred = svc.predict(x_test)

            conf_mat = confusion_matrix(y_test, y_pred)
            tem = conf_mat.ravel() #ravel mengembalikan contigous array
            acc = (tem[3]+tem[0])/(tem[3]+tem[0]+tem[1]+tem[2])*100
            print("Fold-", q, "\nNilai C :", c[i], "\nNilai Gamma :", gm[j], "\nAccuracy :", acc, "\n")
        #     precision = (tem[3]/(tem[3]+tem[1]))*100
        #     recall = (tem[3]/(tem[3]+tem[2]))*100

#             print("Fold-", i, "\nAccuracy :", acc, "\nPrecission :", precision, "\nRecall :", recall, "\n")

    q+=1
```
